```python
import requests
from bs4 import BeautifulSoup
```

```python
# from requests.api import get

def getLinks(url):
  try:
    response = requests.get(url, timeout = 3)
  except:
    # print('some error occured')
    return []
  links_needed = []

  soup = BeautifulSoup(response.content, 'html.parser')
  links = soup.find_all('a')
  get_only = ('.in','.com','.in/','.com/')

  for link in links:
    link = link.get('href')
    if link:
      if(str(link).endswith('/')):
        link = link.rstrip(link[-1])                 #filter for only .in,.com /.in,/.com
      if link in links_needed or link==url:
        continue
      # print(link)
      if '@' in link:
        continue
      if(str(link).endswith(get_only)):
        links_needed.append(link)

  return links_needed

def checkTarget(url,links):
  for link in links:                 # check target url
    if url == link:
      return url
  return False

url = 'https://iiitk.ac.in'                  # start node or Initial url
graph = {}
graph[url] = getLinks(url)
visited = []
queue = []
loop_count = 0
limit = 4                                #levels checking till 3 levels(including root)
visited.append(url)
queue.append(url)

while queue:
  m = queue.pop(0)
  #print (m, end = " ")
  loop_count += 1
  #print(loop_count)
  if(m in graph.keys()):
    for neighbour in graph[m]:
      if neighbour not in visited:
        visited.append(neighbour)
        queue.append(neighbour)
        if loop_count< limit:
          graph[neighbour] = getLinks(neighbour)
```

```python
for key in graph.keys():
  print(key,"->>>",graph[key])
```

```python
def find_path( start_vertex, end_vertex, path=None):
        """Find a path from start_vertex to end_vertex in graph."""
        if path == None:
            path = []

        raph = graph

        path = path + [start_vertex]
        if start_vertex.__eq__(end_vertex):
            return path
        if start_vertex not in raph:
            return None
        for vertex in raph[start_vertex]:
            if vertex not in path:
                extended_path = find_path(vertex, end_vertex, path)
                if extended_path:
                    return extended_path
        return None

goal = 'http://digitalindia.gov.in'
if(str(goal).endswith('/')):
      goal = goal.rstrip(goal[-1])
print(find_path(url, goal))
```
